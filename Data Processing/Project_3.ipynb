```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
q1 = pd.read_csv('anuraagr_si618_proj_output2.csv', header = None)
```

```python
new_q1 = q1[1:len(q1) - 1]
new_q1 = new_q1.rename(columns={0: 'Year', 1: '% movies social'})
```

```python
new_q1 = new_q1.sort_values(by = 'Year')
```

```python
plt.bar(new_q1.Year, new_q1["% movies social"],width = 1)
plt.title("Socially aware movies per year")
plt.xlabel("Years")
plt.ylabel("Percentage of aware")
plt.xticks(new_q1.Year[:len(new_q1):10])
```
